```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Řídké náhodné grafy

Vygenerován graf na s n vrcholy a cn hranami, c = 1,...,15

```python
sparse_constants = range(1,16)

def labels(axis):
    axis.set_xlabel("počet vrcholů")
    axis.set_ylabel("čas (ms)")
    
fig,axes = plt.subplots(nrows=len(sparse_constants),ncols=2,figsize=(20,100))

runs = []

for i,c in enumerate(sparse_constants):
    df = pd.read_csv("../out/sparse{c}.csv".format(c=c))
    runs.append((c,len(df), len(df[df['lp']<df['ek']])))
    labels(axes[i][0])
    labels(axes[i][1])
    axes[i][0].set_title("LP, c = {c}".format(c=c))
    axes[i][1].set_title("EDMONDS KARP, c = {c}".format(c=c))
    axes[i][0].plot(df['n'],df['lp'])
    axes[i][1].plot(df['n'],df['ek'],color="orange")

for c,r,b in runs:
    print("c={c},runs:{r},ilp better: {b}, percentage: {p:.2f} %".format(c=c,r=r,b=b,p=b/r*100))
    
```
